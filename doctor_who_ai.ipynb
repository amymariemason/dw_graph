```python
def breaklines(name):
    parts = name.split()
    try:
        parts[2] = "\n" + parts[2]
        return(" ".join(parts))
    except IndexError:
        return(name)

fig, ax = plt.subplots(1, 1, figsize=(15, 6))
current_season = 1
count = 0
marker = "+"
for col, doctor in CLASSIC_AI.items():
    if col == 9:
        marker = "s"
    col -= 1
    init_count, doc_min, doc_max = count, 100, 0
    for num, season in doctor.items():
        if num == 20:
            marker = "x"
        elif num == 7 and col > 9:
            marker = "d"
        if num != current_season:
            current_season = num
            ax.vlines(count-0.5, 20, 100, color="0.75", linewidth=1)
        for name, episodes in season.items():
            data = np.ma.masked_equal(episodes, -1)
            if data.min() < doc_min:
                doc_min = data.min()
                doc_min_name = name
            if data.max() > doc_max:
                doc_max = data.max()
                doc_max_name = name
            ax.scatter([count]*len(episodes), data, color=f"C{col}", marker=marker, s=10)
            count += 1
    ax.hlines([doc_min, doc_max], init_count, count, color="k", linewidth=0.5)
    if doc_max_name != doc_min_name:
        ax.text(init_count, doc_min, breaklines(doc_min_name), ha="left", va="top", fontsize=8)
        if doc_max_name.startswith("Rememb"):
            y0 = 78
        elif doc_max_name.startswith("The Parting"):
            y0 = 90
        else:
            y0 = doc_max
        ax.text(init_count, y0, breaklines(doc_max_name), ha="left", va="bottom", fontsize=8)

ax.set_xlabel("Episode number")
ax.set_ylabel("Audience appreciation index")
ax.set_title("Doctor Who Through The Years")
ax.margins(0.0001)
fig.savefig("dw_ai.png", bbox_inches="tight")
plt.show()
```
